```python
# FunciÃ³n para convertir a numÃ©rico o NaN
def convert_to_numeric(value):
    try:
        return pd.to_numeric(value)
    except:
        return np.nan

# Aplicar la funciÃ³n convert_to_numeric a los valores no numÃ©ricos
data = data.applymap(convert_to_numeric)
```
